```python
try:
    import numpy as np
    import matplotlib.pyplot as plt
    import cv2
except Exception as e:
    print("Importing Error: {}".format(e))
```

```python
image = cv2.imread("images/lenna.jpg", cv2.IMREAD_GRAYSCALE)

# Fourier dnmn aln
f_transform = np.fft.fft2(image)

# Kaydrmay (shift) uygulayn
f_transform_shifted = np.fft.fftshift(f_transform)

# Mutlak deeri aln
f_transform_magnitude = np.abs(f_transform_shifted)

# Logaritmay aln
f_transform_log = np.log(f_transform_magnitude + 1)  # +1 eklemek, sfra blme hatasn engeller

# Grnty izdirin
plt.subplot(121), plt.imshow(image, cmap='gray')
plt.title('Giri Grnt'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(f_transform_log, cmap='gray')
plt.title('Fourier Dnm'), plt.xticks([]), plt.yticks([])

plt.show()
```
