```python
try:
    import numpy as np
    import matplotlib.pyplot as plt
    import cv2
except Exception as e:
    print("Importing Error: {}".format(e))
```

```python
image = cv2.imread("images/lenna.jpg", cv2.IMREAD_GRAYSCALE)

# Fourier dönüşümünü alın
f_transform = np.fft.fft2(image)

# Kaydırmayı (shift) uygulayın
f_transform_shifted = np.fft.fftshift(f_transform)

# Mutlak değeri alın
f_transform_magnitude = np.abs(f_transform_shifted)

# Logaritmayı alın
f_transform_log = np.log(f_transform_magnitude + 1)  # +1 eklemek, sıfıra bölme hatasını engeller

# Görüntüyü çizdirin
plt.subplot(121), plt.imshow(image, cmap='gray')
plt.title('Giriş Görüntü'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(f_transform_log, cmap='gray')
plt.title('Fourier Dönüşümü'), plt.xticks([]), plt.yticks([])

plt.show()
```
