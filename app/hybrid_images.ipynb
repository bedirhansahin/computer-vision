```python
try:
    import cv2
    import numpy as np
    import matplotlib.pyplot as plt

    print("Libraries Imported")
except Exception as e:
    print(f"Importing error: {e}")
```

```python
dog_image = cv2.imread("../images/dog.bmp")
dog_image_gray = cv2.cvtColor(dog_image, cv2.COLOR_BGR2GRAY)

cat_image = cv2.imread("../images/cat.bmp")
cat_image_gray = cv2.cvtColor(cat_image, cv2.COLOR_BGR2GRAY)


fig, axes = plt.subplots(1, 2, figsize=(6, 3))

# Draw first image
axes[0].imshow(dog_image_gray, cmap="gray")
axes[0].set_aspect("auto")
axes[0].axis("off")

# Draw second image
axes[1].imshow(cat_image_gray, cmap="gray")
axes[1].set_aspect("auto")
axes[1].axis("off")
```

### Gaussian filtered dog (low-pass)


```python
gaussian_filtered_dog = cv2.GaussianBlur(cat_image_gray, (5, 5), 1)

plt.imshow(gaussian_filtered_dog, cmap="gray")
plt.axis("off")
plt.show()
```

### Laplacian Filtered (high-pass)

```python
lablacian_filtered_cat = cv2.Laplacian(dog_image_gray, cv2.CV_16S)

plt.imshow(lablacian_filtered_cat, cmap="gray")
plt.axis("off")
plt.show()
```

### Hybrid Images

```python
hybrid_image = np.uint8(gaussian_filtered_dog + lablacian_filtered_cat)
plt.imshow(hybrid_image, cmap="gray")
plt.axis("off")
plt.show()
```
